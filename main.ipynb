## using minmaxscaler 
MinMaxScaler scales all data features in range [0, 1] or else in range [-1, 1] if there are negative values are present in the dataset. Use it when our data does not follow a normal distribution or when we need scaled data for algorithms like decision trees, k-nearest neighbors or support vector machines. It gives best results when outliers are minimal or absent as it is sensitive to extreme values.

It scales data to a fixed range (typically [0, 1]) by subtracting minimum value and dividing by the range (max - min) which ensures all feature values lies within specified range.
Example: Imagine you're deciding which fruit it is based on its shape and size. You compare it to fruits you already know.

If k = 3, the algorithm looks at the 3 closest fruits to the new one.
If 2 of those 3 fruits are apples and 1 is a banana, the algorithm says the new fruit is an apple because most of its neighbors are apples.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_scaled[0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,random_state=0,test_size = 0.25)
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(x,y,random_state=0,test_size = 0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

## this one is without scaling

```python
model1 = KNeighborsClassifier(n_neighbors=5)
model1.fit(x_train1,y_train1)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_pred1 = model1.predict(x_test1)
y_pred1
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, classification_report
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test1,y_pred1)
```

## Accuracy with feature scaling and without feature scaling

```python
accuracy_score(y_test,y_pred)
```

## acurracy without feature scaling

```python
accuracy_score(y_test1,y_pred1)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(classification_report(y_test1,y_pred1))
```

```python
import matplotlib.pyplot as plt
```

```python
error = []
for k in range(1,41):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    error.append(np.mean(pred != y_test))
```

```python
error
```

```python
plt.figure(figsize=(16,9))
plt.xticks(range(1,41))
plt.grid()
plt.xlabel("value of predictions")
plt.ylabel("value of error")
plt.plot(range(1,41),error, marker='.')
```
